```micropython
%serialconnect to --port=COM3 --baud=115200
import network
import time
import ustruct
import urandom  # Importar urandom para generaci√≥n de n√∫meros aleatorios
from umqtt.simple import MQTTClient

# Configuraci√≥n de la red Wi-Fi
SSID = "TIGO-D26B"
PASSWORD = "4D99ED808110"

# Configuraci√≥n del servidor MQTT Mosquitto
mqtt_server = '192.168.0.5'  # IP de Mosquitto en tu red local
mqtt_port = 1883  # Puerto t√≠pico para Mosquitto sin SSL
mqtt_id = b'ESP8266_client'  # ID √∫nico del cliente
mqtt_topic = b'helion'

# Configuraci√≥n del motor de coordenadas
motor_positions = [0, 45, 90, 135, 180, 225, 270, 315, 360]

def conectar_wifi():
    wlan = network.WLAN(network.STA_IF)
    wlan.active(True)

    if not wlan.isconnected():
        print('Conectando a la red Wi-Fi...')
        wlan.connect(SSID, PASSWORD)

    while not wlan.isconnected():
        time.sleep(1)

    print('Conexi√≥n Wi-Fi establecida:', wlan.ifconfig())

def simular_datos():
    # Simulaci√≥n del sensor LDR (promedio de 300, con algo de variaci√≥n)
    ldr_value = urandom.getrandbits(8) % 21 + 290  # Genera un n√∫mero aleatorio entre 290 y 310

    # Simulaci√≥n del sensor de temperatura (promedio de 25.25, con variaci√≥n entre 24 y 26)
    temperature_value = round(urandom.getrandbits(8) % 201 / 10 + 24, 2)  # Genera un n√∫mero aleatorio entre 24.0 y 26.0

    # Simulaci√≥n de la posici√≥n del motor (alternando entre 0 y 360 grados en segmentos de 45)
    motor_position = motor_positions[urandom.getrandbits(8) % len(motor_positions)]  # Elegir aleatoriamente de la lista

    # Combinamos los tres valores en un solo string
    datos = f'{ldr_value},{temperature_value},{motor_position}'
    
    return datos

def main():
    conectar_wifi()
    time.sleep(1)  # Peque√±a espera para estabilidad

    # Configuraci√≥n y conexi√≥n al servidor MQTT sin autenticaci√≥n
    mqtt_client = MQTTClient(mqtt_id, mqtt_server, port=mqtt_port)

    try:
        mqtt_client.connect()
        print('Conectado al servidor MQTT en Mosquitto')

        # Enviar datos continuamente
        while True:
            datos = simular_datos()  # Simulamos los datos de los sensores
            mqtt_client.publish(mqtt_topic, datos.encode('utf-8'))
            print('Datos enviados:', datos)
            time.sleep(3)  # Espera de 3 segundos antes de enviar el siguiente conjunto de datos

    except Exception as e:
        print('Error al conectar o enviar datos:', e)

    finally:
        mqtt_client.disconnect()
        print('Desconectado del servidor MQTT')

main()
```
